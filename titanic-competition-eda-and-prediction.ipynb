‚ñ∂ Lets see which column has how much of missing data.

```python
# Creating a data frame with missing data percentage and number in train dataset
missing_train_col = ['Age','Cabin','Embarked']
missing_train_number = []
missing_train_percentage = []
for col in missing_train_col:
  missing_train_number.append(train_df[col].isnull().sum())
  missing_train_percentage.append(train_df[col].isnull().sum()/train_df.shape[0] * 100)

missing_train_data = pd.DataFrame(
    {'Col_name': missing_train_col,
     'Missing_number': missing_train_number,
     'Missing_percentage': missing_train_percentage}
    )
print("Missing data in train dataset:\n")
print(missing_train_data)


# Creating the same Data frame for test dataset
missing_test_col = ['Age','Fare','Cabin']
missing_test_number = []
missing_test_percentage = []
for col in missing_test_col:
  missing_test_number.append(pre_test_df[col].isnull().sum())
  missing_test_percentage.append(pre_test_df[col].isnull().sum()/pre_test_df.shape[0] * 100)

missing_test_data = pd.DataFrame(
    {'Col_name': missing_test_col,
     'Missing_number': missing_test_number,
     'Missing_percentage': missing_test_percentage}
    )
print("\nMissing data in test dataset:\n")
print(missing_test_data)
```

‚ñ∂ Let's Start with Embarked column from train data because it has only 2 missing rows or about 0.22% of missing data.

‚ñ∂ Because this Embarked data has very less number of missing data we can drops these missing rows.

‚ñ∂ But I will impute these 2 rows with SimpleImputer class from sklearn with 'most_frequent' strategy, So, that we don't lose any data.

```python
embarked_imputer = SimpleImputer(strategy='most_frequent')
train_df['Embarked'] = embarked_imputer.fit_transform(train_df[['Embarked']]).ravel()
train_df['Embarked'].isnull().sum()
```

‚ñ∂ Let's now handle the Fare column from the train data because it has only 1 missing row and has about 0.23% of missing data.

‚ñ∂ Because we will fit the train data into the model to predict the test data we need to fit the train data in Simple imputer to get the mode of that data and transform the test data for better performance of the dataset.

‚ñ∂ As this Fare column have numerical data we can not use Mode/most_frequent strategy, we have to use 'mean' or 'median', Let's decide this by seeing the distribution of this column using kde and histplot

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax1 = sns.histplot(ax=axes[0], data=train_df, x='Fare', kde=True, color='green')
ax1.set_title('Fare distribution in Train Data')
ax1.set_xlabel('Fare')
ax1.set_ylabel('Count')

ax2 = sns.histplot(ax=axes[1], data=pre_test_df, x='Fare', kde=True, color='blue')
ax2.set_title('Fare distribution in Test Data')
ax2.set_xlabel('Fare')
ax2.set_ylabel('Count')

plt.show()
```

‚ñ∂ As we can see the data in Fare column in both Train and test datasets are highly right skewed.So, using mean will not be appropreate. That's why I am using median for strategy for this column.

```python
fare_imputer = SimpleImputer(strategy='median')
fare_imputer.fit(train_df[['Fare']])
pre_test_df['Fare'] = fare_imputer.transform(pre_test_df[['Fare']]).ravel()
print("The number of rows and columns in test data: ", pre_test_df.shape)
pre_test_df['Fare'].isnull().sum()
```

‚ñ∂ Now lets handle the 'Age' column which has 19.86% and 20.57% of missing data in Train and test dataset repectively.

‚ñ∂ Here we can use use different method like simple filling with mean, median and mode for for complex KNN or linear regression.

‚ñ∂ But in this case I am using a group wise imputation by grouping the data with 'Pcalss' and 'Sex' columns because this helps preserve the relation between 'Pclass' , 'Age' and 'Sex' columns, which are likely important for prediction survival.

‚ñ∂ Now, we need to decide which one is better mean or median, that we can decide by seeing the distribution grapgh just like the Fare column.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax1 = sns.histplot(ax=axes[0], data=train_df, x='Age', kde=True, color='green')
ax1.set_title('Age distribution in Train Data')
ax1.set_xlabel('Age')
ax1.set_ylabel('Count')

ax2 = sns.histplot(ax=axes[1], data=pre_test_df, x='Age', kde=True, color='blue')
ax2.set_title('Age distribution in Test Data')
ax2.set_xlabel('Age')
ax2.set_ylabel('Count')

plt.show()
```

‚ñ∂ As per the graph the Age column in both train and test datasets are right skewed we can use median for better imputation.

```python
# For Train Data
train_df['Age'] = train_df.groupby(['Pclass','Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
print("The number of missing values in Age column in Train data: ", train_df['Age'].isnull().sum())

# For Test Data
pre_test_df['Age'] = pre_test_df.groupby(['Pclass','Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
print("The number of missing values in Age column in Test data: ", pre_test_df['Age'].isnull().sum())
```

‚ñ∂ Now we are left with only 'Cabin' column in both train and test datasets. which have approx 78% of missing data in both tha datasets which is a large number to be filled by imputation methods. This can result in inappropreate and bised results.

‚ñ∂ That's by we can't use it in our model for predicting survival of a passenger. But, This column contain a crucial information about the deck of that person.

‚ñ∂ So, I am extracting the deck information from this column and also filling the null values with letter 'U' Unidentified that represent data not available for that row.

```python
# Extrating the DEck From cabin column
# For Train Data
train_df['CabinDeck'] = train_df['Cabin'].astype(str).str[0]
print("The Decks available in Train Data:\n")
print(train_df['CabinDeck'].unique())

# For Test Data
pre_test_df['CabinDeck'] = pre_test_df['Cabin'].astype(str).str[0]
print("\nThe Decks available in Test Data:\n")
print(pre_test_df['CabinDeck'].unique())
```

‚ñ∂ We need to use same type of graph comparison for multiple column so, let create a function for that.

```python
def compare_plot(col):
  fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

  ax1 = sns.countplot(ax=axes[0],x=col, hue='Survived', data=train_df, palette={0: 'salmon', 1: 'skyblue'})
  ax1.set_title(f'{col} vs Survived')
  ax1.set_xlabel(col)
  ax1.set_ylabel('Count')
  handles, labels = ax1.get_legend_handles_labels()
  ax1.legend(handles, ['Did not Survive', 'Survived'], title='Survived')

  crosstab = pd.crosstab(
      train_df[col],
      train_df['Survived'].replace({1: 'Survived', 0: 'Did not Survive'}),
      normalize='index'
  ) * 100

  ax2 = crosstab.plot(
      ax=axes[1],
      kind='bar',
      stacked=True,
      color=['salmon', 'skyblue'])
  ax2.tick_params(axis='x', rotation=0)
  ax2.set_title(f"Bar plot of Crosstab between {col} and Survived")
  ax2.set_xlabel(col) # Added x-label for clarity
  ax2.set_ylabel('Percentage') # Added y-label for clarity

  plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
  plt.show()

  print(f'\nPercentage of survival by {col}:\n{crosstab.round(2)}\n')
```

‚ñ∂ The above function creates two plots first a bar count plot of specified column with the hue of survived column for comparison and the second one with the crosstab bar plot of that column with survived column also for better camparison

‚ñ∂ Lets define a funtion for distribution graph for continuos data

```python
def distribution_plot(col):
  fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

  survived_values = [0, 1]
  colors = ['red', 'green']
  titles = [f'{col} distribution for non-survivors', f'{col} distribution for survivors']

  for ax, survived_value, color, title in zip(axes, survived_values, colors, titles):
      sns.histplot(
          ax=ax,
          data=train_df[train_df['Survived'] == survived_value],
          x=col,
          kde=True,
          bins=20,
          legend=False,
          color=color
      )
      ax.set_title(title)
      ax.grid(True, linestyle='--', alpha=0.5)

  ymin0, ymax0 = axes[0].get_ylim()
  ymin1, ymax1 = axes[1].get_ylim()
  common_ymin = min(ymin0, ymin1)
  common_ymax = max(ymax0, ymax1)

  for ax in axes:
      ax.set_ylim(common_ymin, common_ymax)

  plt.show()
```

<h1> Feature Engineering </h1>

‚ñ∂ Here we will create some new features from the existing one. Remember we have created a 'cabinDeck' column above, Similar to that we will create some more feature to help model understand the data easily for better performance of the model

‚ñ∂ Let's start with 'FamilySize' where total number of family on board will be stored.

```python
train_df['FamilySize'] = train_df['SibSp'] + train_df['Parch'] + 1
print(train_df.head())

pre_test_df['FamilySize'] = pre_test_df['SibSp'] + pre_test_df['Parch'] + 1
print(pre_test_df.head())
```

‚ñ∂ Next one is 'IsAlone' column where we store 0 if the passenger is alone onboard and 1 of passenger is with someone onboard

```python
train_df['IsAlone'] = (train_df['FamilySize'] == 1).astype(int)
print(train_df.head())

pre_test_df['IsAlone'] = (pre_test_df['FamilySize'] == 1).astype(int)
print(pre_test_df.head())
```

‚ñ∂ Next is 'Title' where we will store the title of the passenger like 'Master', 'lady,'Dr', etc. Since these show an hierarchy in the society that's by they can also play an important role in survival of that passenger.

‚ñ∂ Also since we can not use the 'name' column from the dataset this can play an important role in the model.

```python
train_df['Title'] = train_df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
print(train_df['Title'].unique())

pre_test_df['Title'] = pre_test_df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
print(pre_test_df['Title'].unique())
```

‚ñ∂ There are a lot of title. So, let's commbine some of them that falls in high level of hierarchy at that time and name it 'Rare'

```python
train_df['Title'] = train_df['Title'].replace(['Lady','Countess','Capt','Col','Don','Dr','Major','Rev','Sir','Jonkheer','Dona'],'Rare')
print(train_df['Title'].unique())

pre_test_df['Title'] = pre_test_df['Title'].replace(['Lady','Countess','Capt','Col','Don','Dr','Major','Rev','Sir','Jonkheer','Dona'],'Rare')
print(pre_test_df['Title'].unique())
```

‚ñ∂ Here we left with some inconsitent values like:


*   Mile - is also 'Miss' in other language.
*   Ms - can also be called as 'Miss'
*   Mme - can also to called as 'Miss'

lets replace them with 'Miss'



```python
train_df['Title'] = train_df['Title'].replace({'Mile':'Miss','Ms':'Miss','Mme':'Mrs'})
print(train_df['Title'].unique())

pre_test_df['Title'] = pre_test_df['Title'].replace({'Mile':'Miss','Ms':'Miss','Mme':'Mrs'})
print(pre_test_df['Title'].unique())
```

‚ñ∂ That's perfect now we are left with only 6 categories in train Data and 5 in test data.

```python
print(train_df.head())
print(pre_test_df.head())
```
